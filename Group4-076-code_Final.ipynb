```python
def filtering_dataset(dataset, info_dict):
    # Start with the full dataset and filter down based on each condition
    final_df = dataset.copy()  # Create a copy of the dataset to preserve the original

    for key, choice in info_dict.items():
        if key in ('gender', 'category', 'product'):
            #print(f"Filtering by {key}: {choice}")
            final_df = final_df[final_df[key] == choice]  # Filter the dataset progressively
            #print(final_df)
    if info_dict['sustainablty'] == 'Y':
        final_df = final_df[final_df['sustainability'] == 'Sustainable']
    #If the user has inputted no it meand we dont need to do any filtering
            
    return final_df
```

```python
# Function to find similar products based on material and color
def find_similar_products(info_dict, matched_df):
    # Combine material and color as features
    matched_df['features'] = matched_df['materials'] + ' ' + matched_df['color']
    
    # Instantiate the vectorizer
    vectorizer = TfidfVectorizer()
    
    # Fit and transform the dataset features
    tfidf_matrix = vectorizer.fit_transform(matched_df['features'])
    
    # Construct the user's choice from info_dict
    user_choice = info_dict['material'] + ' ' + info_dict['color']
    
    # Transform the user's choice using the fitted vectorizer
    user_choice_vec = vectorizer.transform([user_choice])
    
    # Calculate cosine similarity
    cosine_similarities = cosine_similarity(user_choice_vec, tfidf_matrix).flatten()
    
    # Get the top N similar items (excluding the selected item itself)
    similar_indices = cosine_similarities.argsort()[-5:][::-1]
    
    # Return the top similar items
    return matched_df.iloc[similar_indices]

```

```python
#Sustainability
#If user is sustainability-sensitive, we are going to provide information on environmental impact of products
#Assuming that adidas has a feature to encourage the purchase of green products, we are gooing to provide information on "Green points" that can be collected.

def emission_reduction_msg(data_df):
    print(f"""

    Thank you for expressing your interest in our environmentally-friendly products.
    Here are the details of positive environmental impact that you will make by choosing these products:

    By choosing this {data_df.iloc[0]['name']}, you're helping reduce carbon emissions by 
    approximately {data_df.iloc[0]['emission_reduction']} kg CO2e! And you will collect {data_df.iloc[0]['green_points']} "Green points", 
    which can be applied as a discount for your next "Green" purchase.
    
    At adidas, we're committed to reducing the environmental impact of products we create. 
    Through our rigorous sustainability initiatives, we've implemented advanced tools to measure 
    and reduce the carbon footprint of our products. By purchasing {data_df.iloc[0]['name']}, 
    you're making a significant contribution to our shared goal of a more sustainable future.

    """)
```

```python
#Reading the adidas dataset and cleaning
dataset = pd.read_excel('Adidas_latest.xlsx')
dataset.head()

#cleaning the data set
dataset.rename(columns={'Emission Reduction by Product (kg)': 'emission_reduction',
'Green points':'green_points'}, inplace=True)
dataset.columns = dataset.columns.str.lower().str.strip()
dataset['gender'] = dataset['gender'].str.upper()
dataset['category'] = dataset['category'].str.upper()

#print(dataset[['gender','category','product']])
```

```python
#Main

desired_cols = ['name','gender','category','product','color','selling_price','availability',
'materials','average_rating','sustainability','emission_reduction','green_points']
#dataset = dataset[desired_cols]
#print(dataset[['gender','category','product']])

#We stay in this loop until user decides that they are done with checking our products
while 1:  
    #Will ask the user about their preferences and store it in info_dict
    info_dict = ask_input(dataset)
    #info_dict = {'name': 'm', 'gender':'MEN', 'category': 'CLOTHING', 'product': 'Jacket', 
    #'color': 'Black', 'material': 'Other', 'price_L': 50, 'price_H': 100, 'sustainablty': 'Y'}

    #By calling a function that filters the data set, we will handle showing similar products
    matched_df = filtering_dataset(dataset, info_dict)
    matched_df = pd.DataFrame(matched_df[desired_cols])

    outpu_cols = ['name','product','color','selling_price','availability',
    'materials','average_rating','sustainability']
    
    #Check for exact_match
    exact_match = matched_df[(matched_df['color']==info_dict['color'])&(matched_df['materials']==info_dict['material'])]
    if len(exact_match) > 1:
        #Here we handle the filtering on price range
        price_L = info_dict.get('price_L', 0)  # Default to 0 if price_L is not provided
        price_H = info_dict.get('price_H', float('inf'))  # Default to infinity if price_H is not provided
        price_filtered_df = exact_match[(exact_match['selling_price'] >= price_L) & (exact_match['selling_price'] < price_H)]
        price_filtered_df = pd.DataFrame(price_filtered_df)
        if len(price_filtered_df) != 0:
            last_round_df = price_filtered_df
            print("We found these products for you. They are sorted by popularity:")
            display(HTML("<h2>Matched Products</h2>"))
            display(last_round_df[outpu_cols].sort_values(by = ['availability','average_rating'],ascending=[True, False]))
        else:
            last_round_df = exact_match
            print("Unfortunately we couldn't find the best option based on your price range")
            print("However, here are some alternative products, They are sorted by popularity:")
            display(HTML("<h2>Alternative Products</h2>"))
            display(last_round_df[outpu_cols].sort_values(by = ['availability','average_rating'],ascending=[True, False]))
    else:
        #Find similar products by applying cosine similarity on color and material
        similar_products = find_similar_products(info_dict, matched_df)
        
        #Here we handle the filtering on price range
        price_L = info_dict.get('price_L', 0)  # Default to 0 if price_L is not provided
        price_H = info_dict.get('price_H', float('inf'))  # Default to infinity if price_H is not provided
        price_filtered_df = similar_products[(similar_products['selling_price'] >= price_L) & (similar_products['selling_price'] < price_H)]
        price_filtered_df = pd.DataFrame(price_filtered_df)
        if len(price_filtered_df) != 0:
            last_round_df = price_filtered_df
            print('''We didn't find an exact match but we recommend these similar products. 
            They are sorted by popularity:''')
            display(HTML("<h2>Similar Products</h2>"))
            display(last_round_df[outpu_cols].sort_values(by = ['availability','average_rating'],ascending=[True, False]))
        else:
            last_round_df = similar_products
            print("Unfortunately we couldn't find the best option based on your price range")
            print("Please check these similar products. They are sorted by popularity:")
            display(HTML("<h2>Alternative Similar Products</h2>"))
            display(last_round_df[outpu_cols].sort_values(by = ['availability','average_rating'],ascending=[True, False]))

    #If the user had prefernces on sustainability, they will see a message on the positive impacts
    if info_dict['sustainablty'] == 'Y':
        emission_reduction_msg(last_round_df)

    #Ask the user if they want to see the questions from beginning
    exit = input("Do you want to check our other products? Y:Yes/N:No").upper()
    if exit == "Y" or exit == "YES":
        continue
    else:
        break

print("Thanks for using Addy! Hope to see you again soon.")
#print(info_dict)
```
