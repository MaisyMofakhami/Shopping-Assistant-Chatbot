```python
from IPython.display import display

import pandas as pd   
dataset = pd.read_csv('adidas_dataset.csv')
dataset.head()
```

```python
info_dict = {}
#Greetings
info_dict['name'] = input("Hi I'm Addy from Adidas, what should I call you?")
print(f"Welcome {info_dict['name']}")

#Asking about the gender and we stay in this loop until we get a valid answer
info_dict['Gender'] = input("Are you interested in Men/Women/Kids products? (M:Men, W:Women, K:Kids)").upper()
while 1:
    if info_dict['Gender'] == 'W':
        info_dict['Gender'] = 'WOMEN'
    elif info_dict['Gender'] == 'M':
        info_dict['Gender'] = 'MEN'
    elif info_dict['Gender'] == 'K':
        info_dict['Gender'] = 'KIDS'
    if info_dict['Gender'] in ('WOMEN','MEN','KIDS'):
        break
    else:
        info_dict['Gender'] = input("Invalid input! Please choose from the given options.(M:Men, W:Women, K:Kids)").upper()
```

```python
while 1:
    #In this loop we will get a valid input on category from user
    while 1:
        info_dict['category'] = input("What type of product are you looking for: (C:Clothing, S:Shoes or A:Accessories?)").upper()
        if info_dict['category'] == 'C':
            info_dict['category'] = 'CLOTHING'
        elif info_dict['category'] == 'S':
            info_dict['category'] = 'SHOES'
        elif info_dict['category'] == 'A':
            info_dict['category'] = 'ACCESSORIES'
        #We stay here until we get valid input
        if info_dict['category'] in ('CLOTHING','SHOES','ACCESSORIES'):
             break
        else:
            info_dict['category'] = input("Invalid input! Please choose from the given options:(C:Clothing, S:Shoes or A:Accessories?)").upper()
    
    #Clothing
    if info_dict['category'] == 'CLOTHING':

        #Here we will about the subcategory or product that the user wants 
        #But based on the producs we actually have in this special category)
        print("We have the following products in this category.")
        #Filter the dataframe
        Cat_product_types = dataset.query("category=='Clothing'")
        # Select the 'product' column and reset the index
        Cat_product_types = Cat_product_types[['Product']].drop_duplicates().reset_index(drop=True)
        #Print the list of options
        display(Cat_product_types)
        #We stay in this loop until we get a valid number
        #In these kind of loops we check both the validity of type of input and its value
        while 1:
            try:   
                selected_no = input("Which one are you interested in? Choose from above numbers:")
                if int(selected_no) in (Cat_product_types.index.tolist()):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        # Get the product name based on the user's input
        info_dict['type'] = Cat_product_types.loc[int(selected_no), 'Product']
        #print(info_dict)

        #Asking about the color that the user wants (only show available colors in this category)
        print("We have the following colors in this category.")
        #Filter the dataframe
        Cat_colors = dataset.query("category=='Clothing'")
        # Select the 'color' column and reset the index
        Cat_colors = Cat_colors[['color']].drop_duplicates().reset_index(drop=True)
        #Print the list of options
        display(Cat_colors)
        #We stay in this loop until we get a valid number
        while 1:
            try:
                selected_no = input("Which one are you interested in? Choose from above numbers:")
                if int(selected_no) in (Cat_colors.index.tolist()):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        # Get the color based on the user's input
        info_dict['color'] = Cat_colors.loc[int(selected_no), 'color']
        #print(info_dict)

        #Shoes
    if info_dict['category'] == 'SHOES':

        #Asking about the subcategory or product that the user wants
        print("We have the following products in this category.")
        #Filter the dataframe
        Cat_product_types = dataset.query("category=='Shoes'")
        # Select the 'product' column and reset the index
        Cat_product_types = Cat_product_types[['Product']].drop_duplicates().reset_index(drop=True)
        #Print the list of options
        display(Cat_product_types)
        #We stay in this loop until we get a valid number
        while 1:
            try: 
                selected_no = input("Which one are you interested in? Choose from above numbers:")
                if int(selected_no) in (Cat_product_types.index.tolist()):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        # Get the product name based on the user's input 
        info_dict['type'] = Cat_product_types.loc[int(selected_no), 'Product']
        #print(info_dict)

        #Asking about the color that the user wants (only show available colors in this category)
        print("We have the following colors in this category.")
        #Filter the dataframe
        Cat_colors = dataset.query("category=='Shoes'")
        # Select the 'color' column and reset the index
        Cat_colors = Cat_colors[['color']].drop_duplicates().reset_index(drop=True)
        #Print the list of options
        display(Cat_colors)
        #We stay in this loop until we get a valid input
        while 1:
            try: 
                selected_no = input("Which one are you interested in? Choose from above numbers:")
                if int(selected_no) in (Cat_colors.index.tolist()):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        # Get the color based on the user's input
        info_dict['color'] = Cat_colors.loc[int(selected_no), 'color']
        #print(info_dict)

        #Accessories
    if info_dict['category'] == 'ACCESSORIES':

        #Asking about the subcategory or product that the user wants
        print("We have the following products in this category.")
        #Filter the dataframe
        Cat_product_types = dataset.query("category=='Accessories'")
        # Select the 'product' column and reset the index
        Cat_product_types = Cat_product_types[['Product']].drop_duplicates().reset_index(drop=True)
        #Print the list of options
        display(Cat_product_types)
        #We stay in this loop until we get a valid number
        while 1:
            try: 
                selected_no = input("Which one are you interested in? Choose from above numbers:")
                if int(selected_no) in (Cat_product_types.index.tolist()):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        # Get the product name based on the user's input
        info_dict['type'] = Cat_product_types.loc[int(selected_no), 'Product']
        #print(info_dict)

        #Asking about the color that the user wants (only show available colors in this category)
        print("We have the following colors in this category.")
        #Filter the dataframe
        Cat_colors = dataset.query("category=='Accessories'")
        # Select the 'color' column and reset the index
        Cat_colors = Cat_colors[['color']].drop_duplicates().reset_index(drop=True)
        #Print the list of options
        display(Cat_colors)
        #We stay in this loop until we get a valid input
        while 1:
            try: 
                selected_no = input("Which one are you interested in? Choose from above numbers:")
                if int(selected_no) in (Cat_colors.index.tolist()):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")

        # Get the color based on the user's input
        info_dict['color'] = Cat_colors.loc[int(selected_no), 'color']

    #We stay in this loop until we get a valid input for the budget
    while 1:
        select_the_budget = input("Please specify your budget range: 0:$0-$49, 1:$50-$100 or 2:More than $100")
        if select_the_budget=="0":
            info_dict["price"]="0-49"
            break
        elif select_the_budget=="1":
            info_dict["price"]="50-100"
            break
        elif select_the_budget=="2":
            info_dict["price"]="More than 100"
            break
        else:
            print("Invalid input! Please try again")

    #Here we ask about the user prefernces on sustainalble products
    while 1:
            try: 
                info_dict['sustainablty'] = int(input("Do you have any preference for sustainable products? ðŸŒ¿ \n0:No, I don't.\n1:Yes, I prefer sustainable options.\n2:Absolutely, I only purchase sustainable products."))
                if info_dict['sustainablty'] in (0,1,2):
                    break
                else:
                    print("Invalid number! Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    #This part is for handling showing similar products which will be completed in Assignment2
    Is_matched = 1 #for example
    Is_available = 1 #for example
    if Is_matched:
        if Is_available:
            print("We recommend these products to you. They are sorted by popularity:")
        else:
            print("The best product for you is currently out of stuck. We will keep ypu posted.")
            print("Meanwhile, you may want to see these similar products:")
    else:
        print("Unfortunately we couldn't find the best option based on your preference")
        print("Also, please check these similar products that we found for you:")

    #Ask the user if they want to see the questions from beginning
    exit = input("Do you want to want to check our other products? Y:Yes/N:No").upper()
    if exit == "Y" or exit == "YES":
        continue
    else:
        break

print("Thanks for using Addy! Hope to see you again soon.")
#print(info_dict)
```
